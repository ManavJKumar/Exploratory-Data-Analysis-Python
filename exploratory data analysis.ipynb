```python
ax = sns.countplot(x= 'Gender', data= df)

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_gen = df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.barplot(x = 'Gender',y= 'Amount' ,data = sales_gen)
```

### From above graphs we can see that most of the buyers are females and even the purchasing power of females are greater than men

# AGE

```python
ax = sns.countplot(data = df, x = 'Age Group', hue = 'Gender')
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_age = df.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.barplot(x = 'Age Group',y= 'Amount' ,data = sales_age)
```

---

# States 

```python
sales_state = df.groupby(['State'], as_index= False)['Orders'].sum().sort_values(by= 'Orders', ascending = False).head(10)

sns.set(rc={'figure.figsize':(20,7)})
sns.barplot(data = sales_state, x = 'State', y='Orders')
```

### "The graphs demonstrate that Uttar Pradesh, Maharashtra, and Karnataka respectively account for the highest number of orders placed and total sales revenue generated."

---

# Marital Status

```python
ax = sns.countplot(data = df, x = 'Marital_Status')

sns.set(rc={'figure.figsize': (8,5)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_state = df.groupby(['Marital_Status','Gender'],as_index= False)['Amount'].sum().sort_values(by = 'Amount', ascending = False)
sns.set(rc={'figure.figsize':(6,5)})
sns.barplot(data= sales_state, x='Marital_Status', y='Amount', hue= 'Gender')
```

### "The visualizations indicate that a majority of the consumers making purchases are wed women who possess substantial spending capacity."

---

# Occupation

```python
sns.set(rc={'figure.figsize':(23,6)})
ax = sns.countplot(data = df, x= 'Occupation')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_state = df.groupby(['Occupation'], as_index= False)['Amount'].sum().sort_values(by='Amount', ascending = False)

sns.set(rc={'figure.figsize':(23,6)})
sns.barplot(data = sales_state, x ='Occupation', y = 'Amount')
```

### "The graphs reveal that a sizable portion of consumers making purchases are employed in the information technology, healthcare, and aviation industries."

---

# Product Category

```python
sns.set(rc={'figure.figsize':(25,6)})
ax = sns.countplot(data = df, x='Product_Category')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
state_sales = df.groupby(['Product_Category'], as_index= False)['Amount'].sum().sort_values(by = 'Amount', ascending= False).head(10)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data= state_sales, x='Product_Category', y='Amount')
```

### "The graphs indicate that the majority of items sold belong to the food, apparel, and electronics product categories."

```python
state_sales = df.groupby(['Product_ID'], as_index= False)['Orders'].sum().sort_values(by = 'Orders', ascending= False).head(10)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data= state_sales, x='Product_ID', y='Orders')
```

```python
fig1, ax1 = plt.subplots(figsize=(10,7))
df.groupby('Product_ID')['Orders'].sum().nlargest(10).sort_values(ascending=False).plot(kind='bar')
```

# Conclusion
